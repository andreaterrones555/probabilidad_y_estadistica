```
from sklearn.linear_model import LinearRegression
from scipy.stats import pearsonr
import numpy as np

# Función para ajustar regresión lineal y calcular correlación
def regresion_lineal(df, nombre):
    print(f"\n--- {nombre} ---")

    X = df[['HorasEstudio']].values  # variable independiente
    y = df['Calificacion'].values    # variable dependiente

    # Modelo de regresión lineal
    modelo = LinearRegression()
    modelo.fit(X, y)

    # Obtener pendiente (coeficiente) e intercepto
    pendiente = modelo.coef_[0]
    intercepto = modelo.intercept_
    print(f" Ecuación de la línea: y = {pendiente:.4f}x + {intercepto:.4f}")

    # Coeficiente de correlación de Pearson
    r, _ = pearsonr(df['HorasEstudio'], df['Calificacion'])
    print(f" Coeficiente de correlación de Pearson: r = {r:.4f}")

# Aplicar a los tres grupos
regresion_lineal(df1, "Grupo 1")
regresion_lineal(df2, "Grupo 2")
regresion_lineal(df3, "Grupo 3")
```

```
import matplotlib.pyplot as plt
import numpy as np

# Función para graficar datos y línea de regresión
def graficar_resultados(df, nombre):
    print(f"\n--- Visualización de {nombre} ---")

    X = df[['HorasEstudio']].values  # variable independiente
    y = df['Calificacion'].values    # variable dependiente

    # Entrenar el modelo de regresión lineal (si no lo hicimos previamente)
    modelo = LinearRegression()
    modelo.fit(X, y)

    # Predicción para la línea de regresión
    y_pred = modelo.predict(X)

    # Crear el gráfico de dispersión (scatter plot)
    plt.figure(figsize=(8, 6))
    plt.scatter(df['HorasEstudio'], df['Calificacion'], color='blue', label='Datos reales', alpha=0.7)

    # Graficar la línea de regresión
    plt.plot(df['HorasEstudio'], y_pred, color='red', label='Línea de regresión', linewidth=2)

    # Añadir título y etiquetas
    plt.title(f'Gráfico de dispersión y línea de regresión - {nombre}')
    plt.xlabel('Horas de Estudio')
    plt.ylabel('Calificación')
    plt.legend()
    plt.grid(True)

    # Mostrar el gráfico
    plt.show()

# Graficar para cada grupo
graficar_resultados(df1, "Grupo 1")
graficar_resultados(df2, "Grupo 2")
graficar_resultados(df3, "Grupo 3")
```

```
from scipy.stats import pearsonr

# Función para calcular el coeficiente de correlación de Pearson
def calcular_correlacion(df, nombre):
    print(f"\n--- Correlación de Pearson - {nombre} ---")

    # Calcular coeficiente de correlación de Pearson
    correlacion, _ = pearsonr(df['HorasEstudio'], df['Calificacion'])
    print(f"Coeficiente de correlación de Pearson (r): {correlacion:.4f}")

# Calcular correlación para cada grupo
calcular_correlacion(df1, "Grupo 1")
calcular_correlacion(df2, "Grupo 2")
calcular_correlacion(df3, "Grupo 3")
```



#### 1. **Coeficiente de regresión (pendiente)**

El coeficiente de regresión representa el cambio promedio en la calificación por cada hora adicional de estudio. Si la pendiente es positiva, como en nuestros casos, significa que a mayor tiempo de estudio, mayor será la calificación obtenida. Por ejemplo, una pendiente de 2.5 indica que por cada hora adicional, las calificaciones aumentan aproximadamente 2.5 puntos.

#### 2. **Coeficiente de correlación (r)**

El coeficiente de correlación de Pearson mide la fuerza y dirección de la relación entre las horas de estudio y las calificaciones:

* **r > 0**: Relación positiva (más estudio, mejores calificaciones).
* **r cercano a 1**: Relación positiva fuerte.
* **r cercano a 0**: Poca o ninguna correlación.
  En este análisis, valores altos de r (cercanos a 1) indican una fuerte relación directa entre ambas variables.

#### 3. **Significancia del modelo**

El modelo es estadísticamente significativo si el **valor p < 0.05**, lo que indica que existe una relación real entre el tiempo de estudio y las calificaciones, y que no es producto del azar. En nuestros resultados, los modelos para el Grupo 2 y Grupo 3 cumplen este criterio.

#### 4. **Confiabilidad del modelo (R²)**

El coeficiente de determinación R² indica qué porcentaje de la variación en las calificaciones puede explicarse por las horas de estudio:

* **R² cercano a 1**: Alta capacidad predictiva.
* **R² bajo**: Modelo poco confiable.
  Cuando R² es alto (mayor a 0.8), se puede confiar en el modelo para hacer predicciones razonables.

## Resumen

* **Pendiente (coef. regresión)**: Muestra cuánto aumentan las calificaciones por cada hora adicional de estudio.
* **Coef. de correlación (r)**: Indica la dirección y fuerza de la relación entre ambas variables.
* **Valor p**: Nos dice si la relación es estadísticamente significativa.
* **R²**: Evalúa qué tan bien el modelo explica los datos
